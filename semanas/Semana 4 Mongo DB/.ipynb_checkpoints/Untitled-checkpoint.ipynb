```python
import time
from pymongo import MongoClient

# 1. Configuración de conexión
try:
    client = MongoClient('database', 27017)
    db = client['proyecto_bigdata']
    coleccion = db['datos_libros'] 
    print("CONEXIÓN EXITOSA: MongoDB está listo.")
except Exception as e:
    print("ERROR DE CONEXIÓN:", e)

# 2. Configuración de tu grupo
NOMBRE_GRUPO = "Mercado_Laboral_TI"
CATEGORIA_ACTUAL = "Tecnologia"

# 3. Datos de prueba (Cargos y Sueldos)
ofertas_ti = [
    {"titulo": "Data Engineer Senior", "precio": "3500000"},
    {"titulo": "Desarrollador Fullstack React", "precio": "1800000"},
    {"titulo": "Especialista Cloud AWS", "precio": "2800000"},
    {"titulo": "Analista QA", "precio": "1500000"},
    {"titulo": "Arquitecto de Software", "precio": "4200000"}
]

# 4. Guardar en MongoDB
for oferta in ofertas_ti:
    registro = {
        "item": oferta["titulo"],
        "valor": float(oferta["precio"]),
        "categoria": CATEGORIA_ACTUAL,
        "grupo": NOMBRE_GRUPO,
        "fecha_captura": time.strftime("%Y-%m-%d %H:%M:%S")
    }
    coleccion.insert_one(registro)

print(f"PROCESO FINALIZADO: Se han guardado {len(ofertas_ti)} ofertas.")
```
