```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager

# 1. Configuración de opciones para que funcione en Docker
options = Options()
options.add_argument("--headless") # Ejecución sin ventana (necesario en Docker)
options.add_argument("--no-sandbox") # Bypass de seguridad del OS
options.add_argument("--disable-dev-shm-usage") # Evita errores de memoria compartida
options.add_argument("--disable-gpu") # Mejora estabilidad

try:
    # 2. El Manager instala el driver compatible automáticamente
    print("Iniciando configuración del Driver...")
    service = Service(ChromeDriverManager().install())
    driver = webdriver.Chrome(service=service, options=options)
    
    # 3. Prueba de conexión a Google
    driver.get("https://www.google.com")
    print(f"¡CONECTADO! Título de la página: {driver.title}")
    
    # Cerramos el navegador
    driver.quit()
    print("Prueba finalizada con éxito.")

except Exception as e:
    print(f"Error detectado: {e}")
```
