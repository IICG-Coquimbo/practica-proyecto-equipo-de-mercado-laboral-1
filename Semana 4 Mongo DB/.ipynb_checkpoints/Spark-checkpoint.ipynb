```python
from pyspark.sql import SparkSession
from pyspark.sql import functions as F

# 1. Configuración de la sesión de Spark
# Nota: La URL y el paquete deben ir en una sola línea de texto
spark = SparkSession.builder \
    .appName("Analisis_Categorias") \
    .config("spark.mongodb.read.connection.uri", "mongodb://database:27017/proyecto_bigdata.datos_libros") \
    .config("spark.jars.packages", "org.mongodb.spark:mongo-spark-connector_2.12:10.1.1") \
    .getOrCreate()

# 2. Cargar los datos desde MongoDB
try:
    df = spark.read.format("mongodb").load()
    print("ÉXITO: Conexión establecida. Spark ya puede leer los datos de MongoDB.")
    
    # 3. Verificación rápida (Paso 2 de tu guía)
    df.printSchema()
    df.show(5)
    
except Exception as e:
    print("Hubo un problema al cargar los datos. Revisa que el contenedor 'database' esté corriendo.")
    print(f"Detalle del error: {e}")
```

```python
from pyspark.sql import SparkSession
from pyspark.sql import functions as F

# 1. Configuración de la sesión de Spark (VERSIÓN CORREGIDA)
spark = SparkSession.builder \
    .appName("Analisis_Categorias") \
    .config("spark.mongodb.read.connection.uri", "mongodb://database:27017/proyecto_bigdata.datos_libros") \
    .config("spark.jars.packages", "org.mongodb.spark:mongo-spark-connector_2.12:3.0.1") \
    .getOrCreate()

# 2. Cargar los datos desde MongoDB
try:
    # IMPORTANTE: En versiones anteriores la forma de leer cambia ligeramente
    # Probemos con el formato estándar primero
    df = spark.read.format("com.mongodb.spark.sql.DefaultSource").load()
    
    print("ÉXITO: Conexión establecida con la versión de compatibilidad.")
    df.printSchema()
    df.show(5)
    
except Exception as e:
    print("Error con la versión 3.0.1, intentando formato alternativo...")
    try:
        # Intento 2 si el anterior falla
        df = spark.read.format("mongodb").load()
        df.show(5)
    except:
        print("Error persistente: Es probable que necesites reiniciar el kernel de Jupyter.")
        print(e)
```
